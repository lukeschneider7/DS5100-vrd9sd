```python
# Ex 6.3 
iris = sns.load_dataset('iris')
iris.sort_index(axis=0, ascending=False)
```

```python
# Ex 6.4
counts = iris.species.value_counts()
count_df = counts.to_frame('n')
count_df
```

```python
# Ex 6.5
len(iris[(iris['petal_length'] <= 1.4) | (iris['petal_length'] >= 1.6)])
```

```python
# Ex 6.6 - min columns
iris['mins_column'] = iris.apply(lambda x: min(x.petal_length, x.petal_width), axis=1)
iris.head()
```

```python
# Ex 6.7 - pivot_table
pd.pivot_table(iris, values= ('sepal_width', 'petal_width'),
               columns=['species'], 
               aggfunc = ['mean', 'median'])
```

```python
# Ex 6.8 - concat
df1 = pd.DataFrame({
                    'key': ['jamie', 'bill'],
                    'lval': [14, 22]})

df2 = pd.DataFrame({ 
                   'key': ['jamie', 'bill', 'asher'],
                    'rval': [19, 20, 20]})

inner = pd.merge(df1, df2, on = 'key', how='inner')
inner

```
